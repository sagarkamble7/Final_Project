```python
# feature extraction:
import re

def wordopt(text):
    text = text.lower()
    
    # remove urls:
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    
    # remove html tags:
    text = re.sub(r'<.*?>', '', text)
    
    # remove punctation:
    text = re.sub(r'[^\w\s]', "", text)
    
    # remove digits:
    text = re.sub(r'\d', "", text)
    
    # remove newline characters:
    text = re.sub(r'\n', " ", text)
    
    return text
```
